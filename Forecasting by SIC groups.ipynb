```python
Header = ['sic', 'sic_Description', '5YrG', 'count', '1YrG', 'count']
Data = []
sic_list = []

cnx = mysql.connector.connect(user='root', password = 'password', database = 'SEC')
cursor = cnx.cursor()

query = ("Select sic, sic_Description, avg(fiveYR_gr) as FiveYrGrowth, count(fiveYR_gr) as FiveYrcount, "
         "avg(nextYR_gr) as NextYrGrowth, count(nextYR_gr) as NextYrcount "
         "From yahoo_forecast "
         "inner join company_info on company_info.CIK	= yahoo_forecast.CIK "
         "group by company_info.sic "
         "having avg(fiveYR_gr) > 0 and avg(nextYR_gr) > 0 and count(fiveYR_gr)>=3 "
         "order by avg(fiveYR_gr) desc")
cursor.execute(query)
for (sic, sic_Description, FiveYrGrowth, FiveYrcount, NextYrGrowth, NextYrcount) in cursor:
    sic_list.append(sic)
    
    Row = [sic, sic_Description, FiveYrGrowth, FiveYrcount, NextYrGrowth, NextYrcount]
    Data.append(Row)    
print(tabulate(Data, headers = Header))
```
